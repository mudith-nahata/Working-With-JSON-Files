4.**Converting a CSV File to JSON**

```python
df=pd.read_csv('movie_titles_metadata.tsv',sep='\t')
df.to_json('movie_titles_metadata.json')
df
```

 5.**Orient Parameter**

```python
data='{"employee_name": "James", "email": "james@gmail.com", "job_profile": [{"title1":"Team Lead", "title2":"Sr. Developer"}]}'
```

```python
df=pd.read_json(data)
df
```

```python
df=pd.read_json(data,orient="records")
df
```

```python
df=pd.read_json(data,orient="index")
df
```

```python
df=pd.read_json(data,orient="columns")
df
```

```python
kf=pd.DataFrame([['a','b'],['c','d']],
                index=['row1','row2'],
                columns=['col1','col2'])
```

```python
kf.to_json()
```

```python
kf.to_json(orient='index')
kf
```

```python
kf.to_json(orient='columns')
kf
```

```python
kf.to_json(orient='records')
kf
```

```python
kf.to_json(orient='split')
kf
```

```python
kf.to_json(orient='table')
kf
```
